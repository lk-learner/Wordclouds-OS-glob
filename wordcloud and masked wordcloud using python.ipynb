```python
# Create a word cloud image

import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
import os

os.getcwd()
path= r'C:\Users\lavy_\OneDrive\Desktop\Python,ML,Coding'
new_path=os.chdir(path)

# Read the whole text.
text = open(os.path.join('lorem ipsum text wordcloud testing.txt')).read()

wc = WordCloud(background_color="white", max_words=100, 
               stopwords=STOPWORDS, contour_width=3, contour_color='firebrick')

# Generate a wordcloud
wc.generate(text)


# show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Create a word cloud image
# Using a mask you can generate wordclouds in arbitrary shapes.

import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
from PIL import Image
import os

os.getcwd()
path= r'C:\Users\lavy_\OneDrive\Desktop\Python,ML,Coding'
new_path=os.chdir(path)

# Read the whole text.
text = open(os.path.join('lorem ipsum text wordcloud testing.txt')).read()


globe_mask= np.array(Image.open(os.path.join('globe.jpg')))


wc = WordCloud(background_color="white", max_words=100, mask=globe_mask,
               stopwords=STOPWORDS, contour_width=3, contour_color='firebrick')

# Generate a wordcloud
wc.generate(text)

# store to file
wc.to_file(os.path.join("globe1.jpg"))

# show
plt.figure()
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
plt.imshow(globe_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")                     
plt.show()

                     
```
